```python
net_sales_per_month = {}
for j in range(1, 13):
    month_str = '2023-{:02d}'.format(j)
    mask = df['Date'].str.startswith(month_str)
    net_sales = df.loc[mask, 'Total Amount'].sum()
    net_sales_per_month[month_str] = net_sales
print(net_sales_per_month)
```
